### Llama 2에서 사용한 SentnecePiece Tokenizer Load

```python
from sentencepiece import sentencepiece_model_pb2 as sp_pb2_model
import os

new_tokenizer_spm = sp_pb2_model.ModelProto()
new_tokenizer_spm.ParseFromString(open(os.path.join('ko_tokenizer', "tokenizer.model"), "rb").read())
ko_tokenizer_tokenset = set(p.piece for p in new_tokenizer_spm.pieces)
ko_tokenizer_tokenset
```

### Llama 3 Tokenizer Load

```python
from transformers import AutoTokenizer
model = "meta-llama/Meta-Llama-3-8B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
tokenizer.add_tokens(list(ko_tokenizer_tokenset))
```

```python
print(len(tokenizer))
```
